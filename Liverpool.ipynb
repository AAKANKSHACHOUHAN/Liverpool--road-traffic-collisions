Liverpool RTC 

Univariate analysis
1. Most accidents are reported on Fridays, followed by Wednesdays, Tuesdays and thursdays (in that order) Sundays have least reports.Typically working days have a higher number of accidents, and over a sixth of overall accidents occurred on Friday.
2. Most accidents occurred during the typical commuter times (early mornings and end of workday in evenings)
3. Theres a steady decline in no. of reported collisions with the years (2005 has the highest and 2010 has least reports.)
4. 1st Road class has the most reports on type A roads whereas 2nd Road class has most reports on the unclassified type. *( Question : What exactly is 1st class and 2nd class ? )*
5. Most incidents fall into the slight category in terms of severity. Very less (a staggering low ~2%) are fatal. 
6. In most. cases, there are no carriageway hazards reported. A small number accounts for an object ir an animal in the carriageway 
7. In most cases, policemen did attend the scene and for about 19% werent attended (thats a big number). 0.3% were filed under self completion. 
8. Most reported cases has no junction control in place. This variable has a lot of missing data.
